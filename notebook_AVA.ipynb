###### Definindo Função

```python
def convert_categorical(dataframe, classes=[]):
  for class_ in classes:
      dataframe[class_] = pd.Categorical(dataframe[class_]).codes    
  return dataframe
```

###### Retirando todos os valores ~NAN~ 

As tabelas afetadas são aquelas que contem as notas

```python
df = df.dropna(subset=[ 
        'NU_NOTA_CN',
        'NU_NOTA_CH',
        'NU_NOTA_LC',
        'NU_NOTA_MT'])

df.to_csv('df_nan.csv', index=False)
```

```python
pd.read_csv('df_nan.csv')
```

###### Tratando os valores da camada - CONVERSÃO - 

Convertendo para Numérico todas as notas que serão usadas no calculo

```python
pd.to_numeric(df['NU_NOTA_CN'])
pd.to_numeric(df['NU_NOTA_CH'])
pd.to_numeric(df['NU_NOTA_LC'])
pd.to_numeric(df['NU_NOTA_MT'])
pd.to_numeric(df['NU_NOTA_REDACAO'])
```

###### Ajustando valores maiores que  1000 na coluna NU_NOTA_CN

```python
maiorqmil = df[(df['NU_NOTA_CN'] > 1000)]   
maiorqmil['NU_NOTA_CN'] = maiorqmil['NU_NOTA_CN'] / 10
df_ = df[(df['NU_NOTA_CN'] <= 1000)]   
df = pd.concat([df_, maiorqmil])
```

###### Ajustando valores maiores que  1000 na coluna NU_NOTA_LC

```python
maiorqmil = df[(df['NU_NOTA_LC'] > 1000)]   
maiorqmil['NU_NOTA_LC'] = maiorqmil['NU_NOTA_LC'] / 10
df_ = df[(df['NU_NOTA_LC'] <= 1000)]   
df = pd.concat([df_, maiorqmil])
```

###### Ajustando valores maiores que  1000 na coluna NU_NOTA_MT

###### Criando o campo MEDIA 

```python
df['MEDIA'] = df['NU_NOTA_MT'] + df['NU_NOTA_LC'] + df['NU_NOTA_CH'] + df['NU_NOTA_CN'] + df['NU_NOTA_REDACAO']
df['MEDIA'] = df['MEDIA'] / 5
```

```python
media_MT = df['NU_NOTA_MT'].mean()
media_LC = df['NU_NOTA_LC'].mean()
media_CH = df['NU_NOTA_CH'].mean()
media_CN = df['NU_NOTA_CN'].mean()
media_RE = df['NU_NOTA_REDACAO'].mean()
media_GERAL = df['MEDIA'].mean()
print(f"MEDIA GERAL = {media_GERAL}")
```

```python
RANGE_000_200_publico = df_PUBLICA[(df_PUBLICA['MEDIA'] >= 0 )]
RANGE_000_200_publico = RANGE_000_200_publico[(RANGE_000_200_publico['MEDIA'] < 200)]

RANGE_200_400_publico = df_PUBLICA[(df_PUBLICA['MEDIA'] >= 200 )]
RANGE_200_400_publico = RANGE_200_400_publico[(RANGE_200_400_publico['MEDIA'] < 400)]

RANGE_400_600_publico = df_PUBLICA[df_PUBLICA['MEDIA'] >= 400]
RANGE_400_600_publico = RANGE_400_600_publico[RANGE_400_600_publico['MEDIA'] < 600]

RANGE_600_800_publico = df_PUBLICA[df_PUBLICA['MEDIA'] >= 600]
RANGE_600_800_publico = RANGE_600_800_publico[RANGE_600_800_publico['MEDIA'] < 800]

RANGE_800_1000_publico = df_PUBLICA[df_PUBLICA['MEDIA'] >= 800]
RANGE_800_1000_publico = RANGE_800_1000_publico[RANGE_800_1000_publico['MEDIA'] < 1000]
```

```python

x1 = round(100*len(RANGE_000_200_privado)/len(df_PRIVADA),2)
x2 = round(100*len(RANGE_200_400_privado)/len(df_PRIVADA),2)
x3 = round(100*len(RANGE_400_600_privado)/len(df_PRIVADA),2)
x4 = round(100*len(RANGE_600_800_privado)/len(df_PRIVADA),2)
x5 = round(100*len(RANGE_800_1000_privado)/len(df_PRIVADA),2)
```

```python
y1 = round(100*len(RANGE_000_200_publico)/len(df_PUBLICA),2)
y2 = round(100*len(RANGE_200_400_publico)/len(df_PUBLICA),2)
y3 = round(100*len(RANGE_400_600_publico)/len(df_PUBLICA),2)
y4 = round(100*len(RANGE_600_800_publico)/len(df_PUBLICA),2)
y5 = round(100*len(RANGE_800_1000_publico)/len(df_PUBLICA),2)
```

```python
labels = ['0-200', '200-400', '400-600', '600-800', '800-1000']
men_means = [x1,x2,x3,x4,x5]
women_means = [y1,y2,y3,y4,y5]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10,8))
rects1 = ax.bar(x - width/2, men_means, width, label='Escolas Privada')
rects2 = ax.bar(x + width/2, women_means, width, label='Escolas Publica')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Porcentagem')
ax.set_title('Porcentagem por faixa de notas')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

#Legenda Setup
ax.legend(wedges, ingredients,
          title="Legenda",
          loc="lower left",
          bbox_to_anchor=(1, 0, 0.5, 1))
        
#Fonte em Geral        
font = {'family' : 'normal',
        #'weight' : 'bold',
        'size'   : 14}
plt.rc('font', **font)
#Tamanho do eixo
plt.axis([-1, 5, 0, 100])

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 10),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
